```python
def sp(x):
    country = ''
    for i in x:
        if i.isalpha() == True or i == ' ':
            country += i
        else:
            return country
    return country


def sb(x):
    country = ''
    for i in x[::-1]:
        if i.isalpha() == True or i == ' ':
            country += i
        else:
            return country[::-1]
    return country[::-1]

def goal(x):
    for i in range(len(x)):
        if x[i].isdigit():
            return x[i:i+3] 
    return goal
```

```python
data['Team1'] = data['highest attendence of Game(s)'].apply(lambda x: sp(x.split(',')[0]))
data['Team2'] = data['highest attendence of Game(s)'].apply(lambda x: sb(x.split(',')[0]))
data['Goal'] = data['highest attendence of Game(s)'].apply(lambda x: goal(x.split(',')[0]))
```

```python
data['Team2'] = data['Team2'].apply(lambda x: x.replace(' ',''))
data['Team1'] = data['Team1'].apply(lambda x: x.replace(' ',''))
```
